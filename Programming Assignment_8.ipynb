```python
def add_matrices():
    l1=[]
    l3=[]
    l4=[]
    try:
        row=int(input("enter the number of rows"))
        col=int(input("enter the number of columns"))
        print("enter data for first matrix row-wise")
        for i in range(row):
            l2=[]
            for k in range(col):
                l2.append(int(input()))
            l1.append(l2)

        print("enter data for second matrix row-wise")  
        for i in range(row):
            l2=[]
            for k in range(col):
                l2.append(int(input()))
            l3.append(l2)

        for i in range(row):
            l2=[]
            for k in range(col):
                l2.append((l1[i][k])+(l3[i][k]))
            l4.append(l2)
    except:
        print("Invalid input")
        
        
    for i in range(row):
        for j in range(col):
            print(l4[i][j],end=" ")
        print()
```

```python
def Multiply_matrices():
    l1=[]
    l3=[]
    l4=[]
    try:
        row=int(input("enter the number of rows"))
        col=int(input("enter the number of columns"))
        print("enter data for first matrix row-wise")
        for i in range(row):
            l2=[]
            for k in range(col):
                l2.append(int(input()))
            l1.append(l2)

        print("enter data for second matrix row-wise")  
        for i in range(row):
            l2=[]
            for k in range(col):
                l2.append(int(input()))
            l3.append(l2)

        for i in range(row):
            l2=[]
            for k in range(col):
                l2.append((l1[i][k])*(l3[i][k]))
            l4.append(l2)
    except:
        print("Invalid input")
        
        
    for i in range(row):
        for j in range(col):
            print(l4[i][j],end=" ")
        print()
```

```python
def transpose_matrices():
    l1=[]
    l3=[]
    row=int(input("enter the number of rows"))
    col=int(input("enter the number of columns"))
    print("enter data for first matrix row-wise")
    for i in range(row):
        l2=[]
        for k in range(col):
            l2.append(int(input()))
        l1.append(l2)
    for i in range(col):
        l2=[]
        for k in range(row):
            l2.append(l1[k][i])
        l3.append(l2)
    for i in range(col):
        for j in range(row):
            print(l3[i][j],end=" ")
        print()
```
